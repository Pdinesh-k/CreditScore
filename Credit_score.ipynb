#Load the DataSet :

```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self,data,label):
    self.data = torch.tensor(data.values).type(torch.float32)
    self.label = torch.tensor(label.values)

  def __len__(self):
    return len(self.data)

  def __getitem__(self,index):
    return self.data[index] , self.label[index]
```

```python
dataset = CustomDataset(x_train , y_train)
```

```python
batch_size = 2
train_loader = DataLoader(dataset , batch_size = batch_size , shuffle = True)
```

```python
class NeuralNetworks(nn.Module):
  def __init__(self,in_features,out_features,hidden_units):
    super().__init__()

    self.layer = nn.Sequential(
    nn.Linear(in_features = in_features , out_features = hidden_units),
    nn.ReLU(),
    nn.Linear(in_features = hidden_units , out_features = hidden_units),
    nn.ReLU(),
    nn.Linear(in_features = hidden_units , out_features = out_features)
    )

  def forward(self,x):
    return self.layer(x)
```

```python
model = NeuralNetworks(in_features = 25 , out_features = 3 , hidden_units = 100)
```

Loss func , Optimizer func , Accuracy func

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params = model.parameters() , lr = 0.01)
```

```python
def accuracy(train_preds , y_train):
  acc = torch.eq(train_preds , y_train).sum().item()
  return acc/len(train_preds)
```

```python
#Random Seed
torch.manual_seed(42)

epochs = 10

for epoch in range(epochs):

  for input , target in train_loader:

    model.train()

    #Forward method
    train_logits = model(input)
    train_preds = torch.softmax(train_logits,1).argmax(dim = 1)
    target = target.view(-1)
    #target = target.type(torch.LongTensor)

    #Calculate the loss
    train_loss = loss_fn(train_logits,target)
    train_accuracy = accuracy(train_preds,target)*100

    #Optimizer
    optimizer.zero_grad

    #LossBackward
    train_loss.backward()

    #Optimizer.step
    optimizer.step()

print(f'epoch : {epoch}\ntrain_loss = {train_loss.item()} , train_accuracy : {train_accuracy}')
```

The accuracy is 100% in Neural Networks !!!!!!!!!!!!!!!!!!!!

```python
test_dataset = CustomDataset(x_train,y_train)
train_loader = DataLoader(test_dataset ,batch_size = 2 , shuffle = True)
```

```python
epochs = 11
model.eval()
with torch.inference_mode():
  for epoch in range(epochs) :
    for inputs,target in train_loader:
      test_logits = model(inputs)
      test_preds = torch.softmax(test_logits,1).argmax(dim=1)
      target = target.view(-1)

      #Loss and accuracy
      test_loss = loss_fn(test_logits,target)
      test_accuracy = accuracy(test_preds , target)*100

print(f'epoch : {epoch}\ntest_loss = {test_loss.item()} , test_accuracy : {test_accuracy}')
```
